```python
from datasets import load_dataset

language = 'zh'
split = 'test'
dataset = load_dataset('Shitao/MLDR', language, split=split, cache_dir='cache_data', trust_remote_code=True)
# save
```

```python
output_data = []
for example in dataset:
    transformed_example = {
        "query": example["query"],
        "pos": [example["positive_passages"][0]['text']],
        "neg": [example["negative_passages"][i]['text'] for i in range(len(example["negative_passages"]))]
    }
    output_data.append(transformed_example)
    
# save to json
import json
with open('transformed_dataset.json', 'w', encoding='utf-8') as f:
    for item in output_data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')
```

```python
test_len = 80
traindataset = output_data[:-test_len]
testdataset = output_data[-test_len:]

# save to json
import json
with open('train_dataset.json', 'w', encoding='utf-8') as f:
    for item in traindataset:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')
with open('test_dataset.json', 'w', encoding='utf-8') as f:
    for item in testdataset:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')
```
